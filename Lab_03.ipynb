```python
table={('Cold','normal'):'Increase',
       ('Cold','malodorous'):'Spray',
       ('Hot','normal'):'Decrease',
       ('Hot','malodorous'):'Spray'}
```

```python
percepts=[]  # to store percept sequence  
def table_driven_agent(percept):
    print('Perception Received: '+ str(percept))
    percepts.append(percept) # updating percept history
    action = lookup(percept,table) # searching for action
    return action

def lookup(p,t):
    return t[p]
```

```python
# lets simulate the agent
import random
Temperature = random.choice(['Cold','Hot'])
Condition = random.choice(['normal','malodorous'])

while True: # to perceieve environment repeatedly
    action = table_driven_agent((Temperature, Condition))
    print('Action Performed: '+ action)
    cmd = input('Get Perception (yes/no): ')
    if(cmd != 'yes'): break
    if action == 'Increase':
        Temperature = 'Hot'
        Condition = random.choice(['normal','malodorous'])
    elif action == 'Decrease':
        Temperature = 'Cold'
        Condition = random.choice(['normal','malodorous'])
    else:
        Condition = 'normal'
```
